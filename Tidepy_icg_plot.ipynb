```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#working directory /media/mohaksharda/MyPassport/Somatic_Cancer_analysis_wo_merging_technical_replicate_bams_17_may_2021/results/icg_tidepy_merged

TIDEPY="tidepy_result_merged_smoke_status_5_6_na_removed.csv"
ICG="merged_transposed_expression_with_smoking_status_w_zscore_1_2only_calculated_for_statistics.tsv"
tidepy = pd.read_csv("%s"%TIDEPY)
icg = pd.read_csv("%s"%ICG,sep="\t")
```

```python
#Extract only those rows with smoking status 1 or 2

tidepy_1 = tidepy[tidepy['Smoking_status'] == 1]
tidepy_2 = tidepy[tidepy['Smoking_status'] == 2]
frames = [tidepy_1, tidepy_2]
tidepy_1_2 = pd.concat(frames)
tidepy_1_2_sort = tidepy_1_2.sort_values(tidepy_1_2.columns[0])
tidepy_1_2_sort
tidepy_1_2_sort.rename(columns = {'Mixture':'TCGA_ID'}, inplace = True)
tidepy_1_2_sort.rename(columns = {'Smoking_status':'smoking_status'}, inplace = True)
tidepy_1_2_sort_indexed = tidepy_1_2_sort.set_index('TCGA_ID')
tidepy_1_2_sort_indexed
```

```python
#Mixture#Extract only those rows with smoking status 1 or 2

icg_1 = icg[icg['smoking_status'] == 1]
icg_2 = icg[icg['smoking_status'] == 2]
frames = [icg_1, icg_2]
icg_1_2 = pd.concat(frames)
icg_1_2_sort = icg_1_2.sort_values(icg_1_2.columns[0])
icg_1_2_sort
icg_1_2_sort.rename(columns = {'Entrez_Gene_Id':'TCGA_ID'}, inplace = True)
icg_1_2_sort_indexed = icg_1_2_sort.set_index('TCGA_ID')
icg_1_2_sort_indexed
```

```python
merged_tidepy_icg = pd.concat([tidepy_1_2_sort_indexed, icg_1_2_sort_indexed], axis=1)
merged_tidepy_icg = merged_tidepy_icg.loc[:,~merged_tidepy_icg.columns.duplicated()]
merged_tidepy_icg
```

```python
merged_tidepy_icg_melted = merged_tidepy_icg.melt ( id_vars = ['No_benefits','Responder','smoking_status','CTL.flag'] , var_name = "Category" , value_name = "Score")
merged_tidepy_icg_melted.head()
```

```python
cell_fraction_boxplot1 = sns.boxplot(y='Score', x='Category', 
                 data=merged_tidepy_icg_melted, 
                 palette="colorblind",
                 hue='smoking_status')

cell_fraction_boxplot1.set_xticklabels(cell_fraction_boxplot.get_xticklabels(), rotation=45, horizontalalignment='right')

cell_fraction_boxplot1.set_xticklabels(cell_fraction_boxplot.get_xticklabels(), rotation=45, horizontalalignment='right')
cell_fraction_boxplot1.set_xlabel("Category", fontsize = 10)
cell_fraction_boxplot1.set_ylabel("Score", fontsize = 10)
#p.set_title("Plot", fontsize = 20)
#plt.legend(labels=["Non-smokers","Smokers"], fontsize = 10)
plt.savefig('tidepy_icg_score_plot.pdf', 
           dpi=300,bbox_inches='tight')
```
