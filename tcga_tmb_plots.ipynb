```python
#tumor_mutational_burden_smoker_non_smokers.csv
#working directory = /media/mohaksharda/MyPassport/Somatic_Cancer_analysis_wo_merging_technical_replicate_bams_17_may_2021/scripts/TCGA_tumor_mutational_burden_analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

TMB_TCGA="tumor_mutational_burden_smoker_non_smokers.csv"

tmb_tcga = pd.read_csv("%s"%TMB_TCGA)
tmb_tcga
```

```python
tmb_tcga_1 = tmb_tcga[tmb_tcga['cohort'] == "non_smokers"]
tmb_tcga_2 = tmb_tcga[tmb_tcga['cohort'] == "smokers"]
frames = [tmb_tcga_1, tmb_tcga_2]
tmb_tcga_1_2 = pd.concat(frames)
tmb_tcga_1_2
```

```python
tmb_boxplot = sns.boxplot(y='total_perMB', x='cohort', 
                 data=tmb_tcga_1_2, 
                 palette="colorblind",
                 hue='cohort')

tmb_boxplot.set_xticklabels(tmb_boxplot.get_xticklabels(), rotation=45, horizontalalignment='right')
tmb_boxplot.set_xticklabels(tmb_boxplot.get_xticklabels(), rotation=45, horizontalalignment='right')
tmb_boxplot.set_xticklabels(tmb_boxplot.get_xticklabels(), rotation=45, horizontalalignment='right')
tmb_boxplot.set_xlabel("Smoking status", fontsize = 10)
tmb_boxplot.set_ylabel("Tumor Mutational Burden (perMB)", fontsize = 10)
#p.set_title("Plot", fontsize = 20)
#plt.legend(labels=["Non-smokers","Smokers"], fontsize = 10)
plt.savefig('tmb_tcga_plot.pdf', 
           dpi=300,bbox_inches='tight')
```

```python
FEATURE_IMP="random_forest_feature_importance_table.csv"
feature_imp = pd.read_csv("%s"%FEATURE_IMP)
feature_imp
```

```python
# Make the PairGrid
g = sns.PairGrid(feature_imp.sort_values("IncMSE", ascending=False),
                 x_vars=['IncMSE'], y_vars=["Factors"],
                 height=10, aspect=.25)

# Draw a dot plot using the stripplot function
g.map(sns.stripplot, size=10, orient="h", jitter=False, palette ="rocket",
      linewidth=1, edgecolor="w")

# Use the same x axis limits on all columns and add better labels
g.set(xlim=(0, 5), xlabel="Feature Importance", ylabel="")

# Use semantically meaningful titles for the columns
titles = ["IncMSE"]

for ax, title in zip(g.axes.flat, titles):

    # Set a different title for each axes
    ax.set(title=title)

    # Make the grid horizontal instead of vertical
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)

sns.despine(left=True, bottom=True)

plt.savefig('incMSE_plot.pdf', 
           dpi=300,bbox_inches='tight')
```
