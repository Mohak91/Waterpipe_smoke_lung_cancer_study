```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

INPUT="/media/dell/MyPassport/Somatic_Cancer_analysis_wo_merging_technical_replicate_bams_17_may_2021/results/14_cibersortx_analysis"
LUAD="luad_tcga_cibersortx_results_merged_smoke_status_5_6_na_removed.csv"

luad = pd.read_csv("%s"%LUAD)
luad
```

```python
#Extract only those rows with smoking status 1 or 2

luad_1 = luad[luad['Smoking_status'] == 1]
luad_2 = luad[luad['Smoking_status'] == 2]
frames = [luad_1, luad_2]
luad_1_2 = pd.concat(frames)
luad_1_2
```

```python
luad_1_2_melted = luad_1_2.melt ( id_vars = ['Mixture','P-value','Correlation','RMSE','Absolute score (sig.score)','Smoking_status'] , var_name = "Cell_type" , value_name = "Cell_fraction")
luad_1_2_melted
```

```python
cell_fraction_boxplot = sns.boxplot(y='Cell_fraction', x='Cell_type',
                 data=luad_1_2_melted, 
                 palette="colorblind",
                 hue='Smoking_status')

cell_fraction_boxplot.set_xticklabels(cell_fraction_boxplot.get_xticklabels(), rotation=45, horizontalalignment='right')
cell_fraction_boxplot.set_xlabel("Cell type", fontsize = 10)
cell_fraction_boxplot.set_ylabel("Cell type abundance", fontsize = 10)
#p.set_title("Plot", fontsize = 20)
#plt.legend(labels=["Non-smokers","Smokers"], fontsize = 10)
plt.savefig('cibersortx_cell_abundance_plot.pdf', 
           dpi=300,bbox_inches='tight')
```
