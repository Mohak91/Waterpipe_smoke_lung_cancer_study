```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#working directory /media/mohaksharda/MyPassport/Somatic_Cancer_analysis_wo_merging_technical_replicate_bams_17_may_2021/scripts/autophagy_analysis
AUTOPHAGY="autophagy_genes_expression_converted_to_zscores.csv"

autophagy = pd.read_csv("%s"%AUTOPHAGY)
autophagy
```

```python
#select top autophagy genes

top_genes= ['BIRC5','BNIP3','CLEC16A','DNM1L','EIF4EBP1','FEZ1','MUL1','PRKD1','RIPK2','SESN1','SESN2','SIRT2','TNFSF10','TRIM22','smoking_status']
autophagy_extracted = autophagy[top_genes]
autophagy_extracted
```

```python
autophagy_extracted_melted = autophagy_extracted.melt ( id_vars = ['smoking_status'] , var_name = "Category" , value_name = "Score")
autophagy_extracted_melted.head()
```

```python
cell_fraction_boxplot1 = sns.boxplot(y='Score', x='Category', 
                 data=autophagy_extracted_melted, 
                 palette="colorblind",
                 hue='smoking_status')

cell_fraction_boxplot1.set_xticklabels(cell_fraction_boxplot.get_xticklabels(), rotation=45, horizontalalignment='right')
cell_fraction_boxplot1.set_xticklabels(cell_fraction_boxplot.get_xticklabels(), rotation=45, horizontalalignment='right')
cell_fraction_boxplot1.set_xlabel("Category", fontsize = 10)
cell_fraction_boxplot1.set_ylabel("Score", fontsize = 10)
#p.set_title("Plot", fontsize = 20)
#plt.legend(labels=["Non-smokers","Smokers"], fontsize = 10)
plt.savefig('autophagy_zscore_plot.pdf', 
           dpi=300,bbox_inches='tight')
```
